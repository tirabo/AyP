## 1. Graficando parejas de datos o funciones

Hablando sin demasiado rigor, si $x$  es un arreglo de NumPy y $f$  es una función definida algebraicamente, obtenemos $y = f(x)$ otro arreglo de NumPy, que está formado por los valores que se obtienen de aplicar $f$ a los valores de $x$. Podemos graficar entonces la función siguiendo el siguiente esquema:
```
y = f(x)
plt.plot(x,y)
plt.show()
```  



Por ejemplo, el siguiente script dibuja el gráfico de la función $y = x^2 + 5$:

````python
from matplotlib import pyplot as plt
import numpy as np

x = np.arange(1,11,0.5)
y =  x**2 + 5
plt.title("Graficando una función con Matplotlib")
plt.xlabel("leyenda del eje x")
plt.ylabel("leyenda del eje y")
plt.plot(x,y)
plt.show()
````

En  el primer renglón se crea un objeto `ndarray` a partir de la función `np.arange()` que serán  los valores del eje `x`. Los valores correspondientes en el eje `y` se almacenan en otro objeto `ndarray`. Estos valores se dibujan utilizando la función `plot()` del submódulo `pyplot` del paquete `matplotlib`.

La representación gráfica se muestra mediante la función `show()`.

Observar que para dibujar o plotear se introduce una "pareja de datos", primero el eje $x$ y  luego los correspondientes valores en el eje $y$. La cantidad de datos en $x$  debe ser igual a la cantidad de datos en $y$.

En  el ploteo hemos agregado texto: un título,  etiqueta del eje $x$ y etiqueta del eje $y$. 

Hay diferentes opciones de ploteo,  que no exploraremos mayormente aquí. Por ejemplo,

````python
from matplotlib import pyplot as plt
import numpy as np

x = np.arange(1,21, 0.5)
y =  2 * x**2 + 3
plt.title("Demostración de Matplotlib")
plt.xlabel("leyenda del eje x")
plt.ylabel("leyenda del eje y")
plt.plot(x,y, "or")
plt.show()
````

En  el código anterior `or` nos dice que se deben dibujar "dots" de color rojo en vez de una curva continua.

También podemos hacer un *gráfico de línea simple* que es una representación visual que muestra cómo varía una variable a lo largo del tiempo o en relación con otra variable. Este tipo de gráfico conecta una serie de puntos mediante líneas. Solo debemos agregar el parámetro `linestyle='-'` al `plot()`:



````python
from matplotlib import pyplot as plt
import numpy as np

x = np.arange(5,21, 3)
y =  2 * x**4 + 3
plt.title("Demostración de Matplotlib")
plt.xlabel("leyenda del eje x")
plt.ylabel("leyenda del eje y")
plt.plot(x,y, linestyle='-')
plt.show()
````

## 2. Trabajando con texto dentro del gráfico
Como vimos al final del punto anterior, existen funciones para añadir texto (etiquetas) a los ejes de la gráfica y a la gráfica en sí. Veamos algunos ejemplos:

````python
from matplotlib import pyplot as plt
import numpy as np

x = np.arange(0, 5, 0.05)
p = plt.plot(x,np.log10(x)*np.sin(x**2)) # log(x)*sen(x**2)
plt.xlabel('Eje X')        # Etiqueta del eje X
plt.ylabel('Eje Y')        # Etiqueta del eje Y
plt.title('Mi grafica')    # Título del gráfico
plt.text(1, -0.4, 'Nota')  # Texto en coodenadas (1, -0.4)
plt.show()
````

En este ejemplo, se usó la función `text()` para añadir un texto arbitrario en la gráfica, cuya posición se debe dar en las unidades de la gráfica. Cuando se utilizan textos también es posible usar fórmulas con formato LaTeX. Veamos un ejemplo,:


````python
from matplotlib import pyplot as plt
import numpy as np

plt.clf()   #limpio la figura
x = np.arange(0.1, 6*np.pi, 0.1)
y = np.sin(x)/x
plt.plot(x, y)
plt.text(2, 0.6, r'$\frac{\operatorname{sen}(x)}{x}$', fontsize=22)
plt.grid()       # Añado una malla al gráfico
plt.title('Representacion de una función')
plt.xlabel('Tiempo / s')
plt.ylabel('Amplitud / cm')
plt.show()
````

Aquí hemos usado código LaTeX para escribir fórmulas matemáticas, para lo que siempre hay que escribir entre `r'$formula$'` y hemos usado un tamaño de letra mayor con el parámetro `fontsize`. En la última línea hemos añadido una grilla con la función `grid()`.

*Nota.* LaTeX es un sistema de escritura orientado a contenidos matemáticos muy popular en matemática,  física teórica y ciencias de la computación.

## 3. Dibujar varias parejas de datos

Se pueden graficar varias funciones en el  mismo ploteo introduciendo varias parejas de datos.

Veamos un ejemplo:

````python
from matplotlib import pyplot as plt
import numpy as np

def f1(x):
    # x es array numpy de números
    y = np.sin(x)
    return y

def f2(x):
    # x es array numpy de números
    y = np.sin(x) + np.sin(5.0*x)
    return y

def f3(x):
    # x es array numpy de números
    y = np.sin(x) * np.exp(-x/10.)
    return y

x = np.arange(0, 10*np.pi, 0.1)


plt.plot(x, f1(x), x, f2(x), x, f3(x))
plt.show()
````

Se pueden poner opciones de ploteo a cada pareja, por ejemplo (aunque no queda muy bien que digamos)

````python
from matplotlib import pyplot as plt
import numpy as np

x = np.arange(0, 10*np.pi, 0.25)
plt.plot(x, f1(x), "or", x, f2(x), x, f3(x), ".b")
plt.show()
````

## 4.  Histogramas
Cuando tenemos un conjunto de datos numéricos, por ejemplo como consecuencia de la medida de una cierta magnitud y queremos representarlos gráficamente para ver la distribución subyacente de los mismos se suelen usar los gráficos llamados *histogramas*.

Los histogramas representan el número de veces que los valores del conjunto caen dentro de un intervalo dado, frente a los diferentes intervalos en los que queramos dividir el conjunto de valores.

En Python podemos hacer histogramas muy fácilmente con la función `hist()` indicando como parámetro un array con los números del conjunto. Si no se indica nada más, se generará un histograma con 10 intervalos (llamados *bins*, en inglés) en los que se divide la diferencia entre el máximo y el mínimo valor del conjunto. Veamos un ejemplo:



````python
nums = np.random.randn(200)  # array con 200 números aleatorios en distribución normal
````

````python
from matplotlib import pyplot as plt
import numpy as np

plt.hist(nums, edgecolor='black')
plt.show()
````


Si no se le proporciona ningún otro argumento a `randn()` produce floats alrededor de $0$ y con una varianza = $1$. Si `n` es entero positivo , entonces `randn(n)` produce un array de `n` valores con floats alrededor de $0$ y con varianza $1$.

Vemos que los números del array se dividieron automáticamente en 10 intervalos (o bins) y cada barra representa para cada una de ellos el número de valores que caen dentro. El parámetro `rwidth` da el ancho de las columnas: con $1$ todas las columna están pegadas, con $0.90$ las columnas están separadas por un 10% del ancho total de la columna llena.  

Si en lugar de usar sólo 10 divisiones queremos usar 15 por ejemplo, debemos indicarlo con el parámetro `bins`:



````python
from matplotlib import pyplot as plt
import numpy as np

plt.hist(nums, bins=15, edgecolor='black')
plt.show()
````

El la figura de abajo se muestra el resultado de superponer ambos histogramas.

````python
from matplotlib import pyplot as plt
import numpy as np

plt.hist(nums, edgecolor='black')
plt.hist(nums, bins=20 , color= 'blue', edgecolor='black')
plt.show()
````

La función `hist()` devuelve una tupla con tres elementos, que son un array con el número elementos en cada intervalo, un array con el punto del eje X donde empieza cada intervalo y una lista con referencias a cada una de las barras para modificar sus propiedades (consulten el manual de matplotlib para encontrar más información y mayores posibilidades de uso).

````python
from matplotlib import pyplot as plt
import numpy as np 

x = plt.hist(nums, bins=20, edgecolor='black')
print('Cantidad de valores del 10º intevalo:',x[0][9]) # cuantos valores están en el 10º intervalo
````

## 5. Gráficos de barras

Los *gráficos de barras* se utilizan para comparar diferentes categorías o variables discretas. Cada barra representa una categoría específica, como tipos de frutas o resultados de encuestas, y la altura de la barra indica la cantidad o frecuencia de cada categoría. Visualmente,  a diferencia del histograma,  se separan las columnas por espacios en blanco.

El submódulo pyplot proporciona la función `bar()` para generar gráficos de barras. El siguiente ejemplo produce el gráfico de barras tomando como parámetros  `x` e `y`.

````python
from matplotlib import pyplot as plt
import numpy as np

x = np.array([5, 6, 8, 9, 10])
y = np.array([10,18, 16,10, 6])
plt.bar(x, y, align = 'center')
plt.title('Gráfico de barras')
plt.ylabel('Eje y')
plt.xlabel('Eje x')
plt.show()
````

## 6. Gráfico circular

Un *gráfico circular*, también conocido como *diagrama circular* o *gráfico de torta*, es una representación gráfica utilizada para mostrar la proporción de diferentes categorías dentro de un todo. Cada sección del círculo representa una categoría y su tamaño es proporcional a la cantidad que representa en relación con el total.

La instrucción que se utiliza para hacer gráficos circulares  es `pie()`

Por ejemplo, mostremos la proporción de diferentes categorías del siguiente conjunto de datos con un gráfico circular.
```
sizes = [20, 30, 25, 25]
labels = ['A', 'B', 'C', 'D']
```

````python
from matplotlib import pyplot as plt
import numpy as np

# Datos
tamaños = [38, 50, 44, 25]
etiquetas = ['A', 'B', 'C', 'D']

# Crear el gráfico circular
plt.pie(tamaños, labels=etiquetas, autopct='%1.1f%%')

# Añadir título
plt.title('Proporción de Diferentes Categorías')

# Mostrar el gráfico
plt.axis('equal')  # Para asegurar que el gráfico sea un círculo
plt.show()
````

Los `labels` son etiquetas externas que identifican cada categoría y la instrucción `autopct='%1.1f%%'`dibuja los porcentajes de cada "porción de la torta". 

## 7. Lectura y escritura de archivos con `numpy`

Los objetos `ndarray` (los array de NumPy) se pueden guardar y cargar. Las funciones de entrada y salida disponibles son

- las funciones `load()` y `save()` manejan archivos binarios numPy (con extensión npy),
- las funciones `loadtxt()` y `savetxt()` manejan archivos de texto.

NumPy introduce un formato de archivo simple para los objetos `ndarray`. El archivo `.npy` almacena los datos, la forma, el `dtype` y otra información requerida para reconstruir el `ndarray`, de manera que el array sea recuperado correctamente aunque el archivo esté en otra máquina con diferente arquitectura.


**numpy.save()**

El  método `save()` almacena el array de entrada en un archivo con extensión `npy`.


````python
import numpy as np

arr_1 = np.array([1,2,3,4,5])
np.save('archivo_salida',arr_1)
````




Para reconstruir el array desde `archivo_salida.npy`, utiliza la función load().


````python
arr_b = np.load('archivo_salida.npy')
print(arr_b)
````



Producirá la siguiente salida

    array([1, 2, 3, 4, 5])

Las funciones `save()` y `load()` aceptan un parámetro booleano adicional   `allow_pickles`. Un pickle en Python se utiliza para serializar y des-serializar objetos antes de guardar o leer de un archivo de disco.



**savetxt()**

El almacenamiento y recuperación de datos de arreglos en formato de texto se realiza con las funciones `savetxt()` y `loadtxt()`.

Ejemplo:

````python
arr_a = np.array([1,2,3,4,5])
np.savetxt('out.txt',arr_a)
arr_b = np.loadtxt('out.txt')
print(arr_b)
````

Producirá la siguiente salida

    [1., 2., 3., 4., 5.]

Las funciones `savetxt()` y `loadtxt()` aceptan parámetros opcionales adicionales como la cabecera, el pie de página y el delimitador.

## 9. La función `subplot()`

La función `subplot()` permite trazar diferentes gráficos en la misma figura, pero como dibujos separados.

En el siguiente script, se grafican el seno y el coseno.

````python
from matplotlib import pyplot as plt
import numpy as np

# Calcular las coordenadas x e y de los puntos de las curvas seno y coseno
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Configurar una grilla tenga ancho 2 y altura 1 (se graficará un plot al costado de otro),
# y establecer la primera de esas subtramas como activa.
plt.subplot(1, 2, 1)

# Se hace el primer plot
plt.plot(x, y_sin)
plt.title('Seno')

# Establecer el segundo subplot como activo y hacer el segundo plot.
plt.subplot(1, 2, 2)
plt.plot(x, y_cos)
plt.title('Coseno')
plt.show()
````

La función `subplot()` de pyplot es utilizada para crear grillas de subplots en una figura de Matplotlib. Esta función acepta *tres* parámetros posicionales: `nrows`, `ncols` e `index`, y opcionalmente otros parámetros para ajustar el tamaño, el espacio y los límites de los subplots. A continuación se describen los tres primeros parámetros de la función `subplot()`:

- `nrows`: número de filas de la grilla de subplots.
- `ncols`: número de columnas de la grilla de subplots.
- `index`: índice del subplot actual, comenzando desde 1 y aumentando de izquierda a derecha y de arriba hacia abajo.

Existesn otros argumentos y parámetros para compartir ejes y para ajustar el tamaño, el espacio y los límites de los subplots.

Es decir el método `subplot()` permite hacer una grilla $m \times n$ y dibujar en cada rectángulo de esa grilla.

````python
from matplotlib import pyplot as plt
import numpy as np

def f1(x):
    y = np.sin(x)
    return y

def f2(x):
    y = np.sin(x) + np.sin(5.0*x)
    return y

def f3(x):
    y = np.sin(x) * np.exp(-x/10.)
    return y

def f4(x):
    y = np.cos(x) * x
    return y

plt.subplot(2, 2, 1)
plt.plot(x, f1(x))
plt.title('sen(x)')

plt.subplot(2, 2, 2)
plt.plot(x, f2(x))
plt.title('sen(x) + sen(5x)')

plt.subplot(2, 2, 3)
plt.plot(x, f3(x))
plt.title('sen(x)*exp(-x/10)')

plt.subplot(2, 2, 4)
plt.plot(x, f4(x))
plt.title('cos(x)*x')

plt.show()
````

No quedó muy bien, pero se puede arreglar con agregando la instrucción `subplots_adjust()` que permite aumentar el espacio vertical:

````python
from matplotlib import pyplot as plt
import numpy as np

def f1(x):
    y = np.sin(x)
    return y

def f2(x):
    y = np.sin(x) + np.sin(5.0*x)
    return y

def f3(x):
    y = np.sin(x) * np.exp(-x/10.)
    return y

def f4(x):
    y = np.cos(x) * x
    return y

plt.subplot(2, 2, 1)
plt.plot(x, f1(x))
plt.title('sen(x)')

plt.subplot(2, 2, 2)
plt.plot(x, f2(x))
plt.title('sen(x) + sen(5x)')

plt.subplot(2, 2, 3)
plt.plot(x, f3(x))
plt.title('sen(x)*exp(-x/10)')

plt.subplot(2, 2, 4)
plt.plot(x, f4(x))
plt.title('cos(x)*x')

# Ajustar el espaciado
plt.subplots_adjust(hspace=0.5)  # Aumenta el espacio vertical

plt.show()
````
